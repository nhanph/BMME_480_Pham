```python
import numpy as np
import pandas as pd
import matplotlib 
import seaborn as sns
from matplotlib import pyplot as plt


dataset=pd.read_csv('winequality-white.csv')

fig,ax=plt.subplots(3,4)

ax[0,0].boxplot(dataset['fixed acidity'])
ax[0,0].set_title('fixed acidity')

ax[0,1].boxplot(dataset['volatile acidity'])
ax[0,1].set_title('volatile acidity')

ax[0,2].boxplot(dataset['citric acid'])
ax[0,2].set_title('citric acid')

ax[0,3].boxplot(dataset['residual sugar'])
ax[0,3].set_title('residual sugar')


ax[1,0].boxplot(dataset['chlorides'])
ax[1,0].set_title('chlorides')

ax[1,1].boxplot(dataset['free sulfur dioxide'])
ax[1,1].set_title('free sulfur dioxide')

ax[1,2].boxplot(dataset['total sulfur dioxide'])
ax[1,2].set_title('total sulfur dioxide')

ax[1,3].boxplot(dataset['density'])
ax[1,3].set_title('density')

ax[2,0].boxplot(dataset['pH'])
ax[2,0].set_title('pH')

ax[2,1].boxplot(dataset['sulphates'])
ax[2,1].set_title('sulphates')

ax[2,2].boxplot(dataset['alcohol'])
ax[2,2].set_title('alcohol')

ax[2,3].boxplot(dataset['quality'])
ax[2,3].set_title('quality')



fig.subplots_adjust(left=0.001, right=0.998, bottom=0.01, top=0.99,
                    hspace=1, wspace=1)

plt.show()
```
